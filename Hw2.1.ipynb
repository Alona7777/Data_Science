```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
print(df)
```

```python
# Вивести перші рядки таблиці за допомогою методу head
df.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df_shape = df.shape
df_shape
```

```python
# Замініть у таблиці значення "—" на значення NaN
df.replace("—", np.nan, inplace=True)
df
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
columns_to_convert = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']  
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum() 
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.drop(index=df.index[-1])
df
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
column_means = df[columns_to_convert].mean()
df.fillna(column_means, inplace=True)
df
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
average_birth_rate_2019 = df['2019'].mean()
regions = df[df['2019'] > average_birth_rate_2019]['Регіон']
regions
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
max_birth_rate_index = df['2014'].idxmax()
max_region = df.loc[max_birth_rate_index]
print(max_region['Регіон'], max_region['2014'])
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.figure(figsize=(10, 8))
plt.bar(df['Регіон'], df['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Народжуваність по регіонах України у 2019 році')
plt.xticks(rotation=90)  # Rotate region names for better readability
plt.tight_layout()  # Adjust layout to fit all labels
plt.show()
```
