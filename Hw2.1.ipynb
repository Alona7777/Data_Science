```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
print(df)
```

```python
# Вивести перші рядки таблиці за допомогою методу head
df.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
df_shape = df.shape
df_shape
```

```python
# Замініть у таблиці значення "—" на значення NaN
df.replace("—", np.nan, inplace=True)
df
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
columns_to_convert = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']  
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum() 
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df = df.drop(index=df.index[-1])
df
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
column_means = df[columns_to_convert].mean()
df.fillna(column_means, inplace=True)
df
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
average_birth_rate_2019 = df['2019'].mean()
regions = df[df['2019'] > average_birth_rate_2019]['Регіон']
regions
```

```python
# У якому регіоні була найвища народжуваність у 2014 році?
max_birth_rate_index = df['2014'].idxmax()
max_region = df.loc[max_birth_rate_index]
print(max_region['Регіон'], max_region['2014'])
```

```python
# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
plt.figure(figsize=(8, 6))
plt.bar(df['Регіон'], df['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Народжуваність по регіонах України у 2019 році')
plt.xticks(rotation=90)  # Rotate region names for better readability
plt.tight_layout()  # Adjust layout to fit all labels
plt.show()
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df = df.apply(pd.to_numeric, errors='coerce')
df.dtypes
print(df)
```

# Visualization

```python
fig, axs = plt.subplots()
axs.plot(df['Регіон'], df['2014'], label='2014')
axs.plot(df['Регіон'], df['2019'], label='2019')
plt.xlabel('Регіон', fontsize='small', color='midnightblue')
plt.ylabel('Народжуваність', fontsize='small', color='midnightblue')
plt.title('Народжуваність по регіонах України у 2019 та 2014 роках', fontsize=15)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.plot(
    df['Регіон'],
    df['1950'],
    label='1950',
    linestyle="--",
    color="#FF5733",
    linewidth=2,
    marker="D"
)
plt.plot(
    df['Регіон'],
    df['2019'],
    label='2019',
    linestyle=":",
    color="#061358",
    linewidth=2,
    marker="*"
)
plt.xlabel('Регіон', fontsize='small', color='midnightblue')
plt.ylabel('Народжуваність', fontsize='small', color='midnightblue')
plt.title('Народжуваність по регіонах України у 1950 та 2019 роках', fontsize=15)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python

plt.pie(
    df['2019'],
    labels=df['Регіон'],
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)

plt.show()
```

```python
df_melted = pd.melt(df, id_vars=['Регіон'], var_name='Рік', value_name='Народжуваність')
sns.set_style("darkgrid")

plt.figure(figsize=(12, 6))
sns.lineplot(x='Регіон', y='Народжуваність', hue='Рік', data=df_melted, marker='o')

plt.title('Народжуваність по регіонах за роками')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_melted, x='Рік', y='Народжуваність', hue='Регіон', palette='Set1', alpha=0.8)
plt.title('Порівняння народжуваності за роками')
plt.xlabel('Рік')
plt.ylabel('Народжуваність')
plt.legend(title='Регіон')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.iloc[:, 1:], palette='Set3')
plt.title('Розподіл народжуваності за роками')
plt.xlabel('Рік')
plt.ylabel('Народжуваність')
plt.show()
```
