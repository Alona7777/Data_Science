```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer
from sklearn.decomposition import PCA
```

```python
folder_path = '/Users/ALONA/Desktop/Go_IT/HW/Data Science/Data_Science/HW_6/data'
dfs = []

for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            dfs.append(df)

df_1 = dfs[0]
df_2 = dfs[1]

df_1.head()
```

```python
df_2.head()
```

### A function for determining the optimal number of clusters using the elbow method

```python
def find_optimal_clusters(data, max_k):
    iters = range(1, max_k + 1)
    sse = []
    for k in iters:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        sse.append(kmeans.inertia_)
    plt.figure(figsize=(10, 6))
    plt.plot(iters, sse, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('SSE')
    plt.show()
```

### Data scaling

```python
scaler = StandardScaler()
df_1_scaled = scaler.fit_transform(df_1)
df_2_scaled = scaler.fit_transform(df_2)
```

### Determination of the optimal number of clusters for each DataFrame

```python
find_optimal_clusters(df_1_scaled, 10)
find_optimal_clusters(df_2_scaled, 30)
```

### yellowbrick

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer_1 = KElbowVisualizer(model, k=(1,12))
visualizer_1.fit(df_1_scaled)
visualizer_1.show()

visualizer_2 = KElbowVisualizer(model, k=(1,12))
visualizer_2.fit(df_2_scaled)
visualizer_2.show()
```

```python
optimal_clusters_df_1 = 2
optimal_clusters_df_2 = 3
```

### Data clustering using K-means

```python
kmeans_df_1 = KMeans(n_clusters=optimal_clusters_df_1, random_state=42)
kmeans_df_1.fit(df_1_scaled)
df_1['Cluster'] = kmeans_df_1.labels_

kmeans_df_2 = KMeans(n_clusters=optimal_clusters_df_2, random_state=42)
kmeans_df_2.fit(df_2_scaled)
df_2['Cluster'] = kmeans_df_2.labels_
```

### Visualization

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df_2_scaled)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=df_2['Cluster'], cmap='viridis')
plt.colorbar(scatter)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Clusters for df_2 with PCA')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_1_scaled[:, 0], df_1_scaled[:, 1], c=df_1['Cluster'], cmap='viridis')
plt.colorbar(scatter)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters for df_1 without PCA')
plt.show()
```
