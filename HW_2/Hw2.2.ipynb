```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
df = pd.read_csv('2017_jun_final - 2017_jun_final.csv')
df
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
df.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
df.isnull().sum()
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
cols_to_delete = ['Спеціалізація', 'Університет', 'Предметна область', 'Валюта', 'cls']
df = df.drop(columns=cols_to_delete, axis=1)
df.head()
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
df.isnull().sum()
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
df = df.dropna()
df.head()
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
df.shape
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = df[df['Мова програмування'] == 'Python']
python_data.head()
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
python_data = python_data.groupby('Посада')
python_data.head()
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

# Aggregating data using the agg method to find min and max salary
agg_python_data = python_data.agg({'Зарплата на місяць': ['min', 'max']})

# Resetting the index for better readability
agg_python_data.reset_index(inplace=True)

# Renaming the columns for better clarity
agg_python_data.columns = ['Position', 'Min Salary', 'Max Salary']
agg_python_data
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    return (row['Min Salary'] + row['Max Salary']) / 12

agg_python_data['avg'] = agg_python_data.apply(fill_avg_salary, axis=1)
agg_python_data
```

```python
# Створіть описову статистику за допомогою методу describe для нового стовпчика.
agg_python_data['avg'].describe()
```

```python
# Збережіть отриману таблицю в CSV файл
agg_python_data.to_csv('avg_salary_python.csv')
```

# Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Зарплата на місяць'], bins=20, kde=True)
plt.title('Distribution of salaries')
plt.xlabel('Salary per month')
plt.ylabel('Value')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.set_context("paper")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Тип компанії', y='Зарплата на місяць')
plt.title('Distribution of salaries by type of company')
plt.xlabel('Type of company')
plt.ylabel('Salary per month')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Вік', y='Зарплата на місяць', hue='Стать', palette='coolwarm')
plt.title('Dependence of salary on age')
plt.xlabel('Age')
plt.ylabel('Salary per month')
plt.legend(title='Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='exp', y='Зміна зарплати за 12 місяців', hue='Стать', marker='s')
plt.title('Salary change for 12 months')
plt.xlabel('Work experience')
plt.ylabel('Salary change')
plt.legend(title='Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Місто', palette='Set2')
plt.title('Number of respondents by city')
plt.xlabel('City')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```
