```python
from ultralytics import YOLO
from google.colab import drive
import os

# model = YOLO('/content/drive/MyDrive/Finel_Project/YOLOv8s_license_plate3/weights/best.pt')




# Подключаем Google Диск
drive.mount('/content/drive')

# Путь к вашему проекту на Google Диске
project_path = '/content/drive/MyDrive/Finel_Project'

# Проверка существования директории, создание если не существует
if not os.path.exists(project_path):
    os.makedirs(project_path)

# Путь к вашему файлу данных и сохранению весов
data_path = '/content/drive/MyDrive/License_Plate_Recognition/data.yaml'
weights_path = os.path.join(project_path, 'YOLOv8s_license_plate3/weights/best.pt')  # Путь к весам предыдущей модели

def train_model():
    # Загрузка модели с использованием весов предыдущего обучения
    model = YOLO(weights_path)

    # Запуск процесса дообучения
    model.train(
        data=data_path,
        imgsz=480,                     # Уменьшенный размер изображений
        epochs=50,                     # Уменьшенное количество эпох
        patience=5,                    # Количество эпох без улучшения для ранней остановки
        name='YOLOv8s_license_plate_finetune',  # Новое имя эксперимента
        project=project_path,          # Путь к проекту на Google Диске
    )

# Вызов функции для начала дообучения
train_model()
```
