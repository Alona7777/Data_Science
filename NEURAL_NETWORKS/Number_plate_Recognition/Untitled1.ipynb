```python
from ultralytics import YOLO
from google.colab import drive
import os

# model = YOLO('/content/drive/MyDrive/Finel_Project/YOLOv8s_license_plate3/weights/best.pt')




# –ü–æ–¥–∫–ª—é—á–∞–µ–º Google –î–∏—Å–∫
drive.mount('/content/drive')

# –ü—É—Ç—å –∫ –≤–∞—à–µ–º—É –ø—Ä–æ–µ–∫—Ç—É –Ω–∞ Google –î–∏—Å–∫–µ
project_path = '/content/drive/MyDrive/Finel_Project'

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏, —Å–æ–∑–¥–∞–Ω–∏–µ –µ—Å–ª–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
if not os.path.exists(project_path):
    os.makedirs(project_path)

# –ü—É—Ç—å –∫ –≤–∞—à–µ–º—É —Ñ–∞–π–ª—É –¥–∞–Ω–Ω—ã—Ö –∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—é –≤–µ—Å–æ–≤
data_path = '/content/drive/MyDrive/License_Plate_Recognition/data.yaml'
weights_path = os.path.join(project_path, 'YOLOv8s_license_plate3/weights/best.pt')  # –ü—É—Ç—å –∫ –≤–µ—Å–∞–º –ø—Ä–µ–¥—ã–¥—É—â–µ–π –º–æ–¥–µ–ª–∏

def train_model():
    # –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –≤–µ—Å–æ–≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –æ–±—É—á–µ–Ω–∏—è
    model = YOLO(weights_path)

    # –ó–∞–ø—É—Å–∫ –ø—Ä–æ—Ü–µ—Å—Å–∞ –¥–æ–æ–±—É—á–µ–Ω–∏—è
    model.train(
        data=data_path,
        imgsz=480,                     # –£–º–µ–Ω—å—à–µ–Ω–Ω—ã–π —Ä–∞–∑–º–µ—Ä –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
        epochs=50,                     # –£–º–µ–Ω—å—à–µ–Ω–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö
        patience=5,                    # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ø–æ—Ö –±–µ–∑ —É–ª—É—á—à–µ–Ω–∏—è –¥–ª—è —Ä–∞–Ω–Ω–µ–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏
        name='YOLOv8s_license_plate_finetune',  # –ù–æ–≤–æ–µ –∏–º—è —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞
        project=project_path,          # –ü—É—Ç—å –∫ –ø—Ä–æ–µ–∫—Ç—É –Ω–∞ Google –î–∏—Å–∫–µ
    )

# –í—ã–∑–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –Ω–∞—á–∞–ª–∞ –¥–æ–æ–±—É—á–µ–Ω–∏—è
train_model()
```
