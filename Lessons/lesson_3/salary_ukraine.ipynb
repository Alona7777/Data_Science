# Data preparation and analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96'
df = pd.read_html(source, match="Середньорічна щомісячна зарплата в Україні з 1992 року", thousands=".", decimal=",")[0]
print(df)
```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96'

tables = pd.read_html(source, thousands=".", decimal=",")

print(f"Found {len(tables)} tables.")

for i, table in enumerate(tables):
    print(f"\nTable {i}:")
    print(table.head())
```

```python
df = tables[2]
df.head()
```

```python
salary_columns = ['year', 'uah', 'usd', 'amount_bread', 'amount_big_mak']
df.columns = salary_columns
df.head()
```

```python
print(df.shape)
print(df.dtypes)
```

```python
df.replace("—", np.nan, inplace=True)
df.replace('н.д.', np.nan, inplace=True)
df = df.replace({'▲': '', '▼': ''}, regex=True)
df = df.replace(r'\[.*?\]', '', regex=True)

df
```

```python
df = df.apply(lambda x: x.str.replace(',', '.').astype(float) if x.dtype == 'object' and x.str.contains(r'^\d+,\d+$').any() else x)
print(df.dtypes)
df
```

```python
df.isnull().sum()
```

```python
df.amount_big_mak = df.amount_big_mak.astype(float)
print(df.dtypes)
```

```python
df.amount_big_mak = df.amount_big_mak.fillna(df.amount_big_mak.mean())
df.head()
```

```python
correlations = df.corr()

print(correlations)
```
